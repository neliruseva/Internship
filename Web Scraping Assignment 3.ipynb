# Q2: In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then scrape all the products available under that product name. Details to be scraped are: 
"Brand
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and
“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
#fetching url to open each product

urls = []
start=0
end=3

for page in range(start,end):
    url=driver.find_elements(By.XPATH, '//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in url[0:180]:
        urls.append(i.get_attribute('href'))
    next_button=driver.find_elements(By.XPATH, '/html/body/div[1]/div[1]/div[1]/div[1]/div/span[1]/div[1]/div[67]/div/div/span/a[3]')
```

```python
len(urls)
```

```python
#creating empty lists and scraping the required data

Brand_Name=[]
Name_of_Product=[]
Price=[]
Return_Exchange=[]
Expected_Delivery=[]
Availability=[]



for i in urls[0:180]:
    driver.get(i)
    time.sleep(3)
    
                                            
    try:
        brand = driver.find_element(By.XPATH,'//span[@class="a-size-base po-break-word"]')
        Brand_Name.append(brand.text)
    except NoSuchElementException:
        Brand_Name.append('-')
        
        
    try:
        product = driver.find_element(By.XPATH, '//span[@id="productTitle"]')
        Name_of_Product.append(product.text)
    except NoSuchElementException:
        Name_of_Product.append('-')
   

    try:
        price = driver.find_element(By.XPATH, '//span[@class="a-price-whole"]')
        Price.append(price.text)
    except NoSuchElementException:
        Price.append('-')
        
        
    try:
        return_exchange = driver.find_element(By.XPATH, '/html/body/div[2]/div/div[5]/div[3]/div[4]/div[24]/div[2]/div/div/div/div[2]/span/div[2]/span')
        Return_Exchange.append(return_exchange.text)
    except NoSuchElementException:
        Return_Exchange.append('-')
        
        
    try:
        expected_delivery = driver.find_element(By.XPATH, '/html/body/div[2]/div/div[5]/div[3]/div[1]/div[3]/div/div[1]/div/div/div/form/div/div/div/div/div[4]/div/div[3]/div[9]/div[1]/div/div/div/span')
        Expected_Delivery.append(expected_delivery.text)
    except NoSuchElementException:
        Expected_Delivery.append('-')
    
    
    try:
        availability = driver.find_element(By.XPATH, '//span[@class="a-size-medium a-color-success"]')
        Availability.append(availability.text)
    except NoSuchElementException:
        Availability.append('-')
        
        
    urls.append(i)
        
```

```python
print(len(Brand_Name), len(Name_of_Product), len(Price), len(Return_Exchange), len(Expected_Delivery), len(Availability))
```

```python
#creating a data frame

df = pd.DataFrame({"Brand Name":Brand_Name, "Name of Product":Name_of_Product, "Price":Price,"Return/Exchange":Return_Exchange, "Expected Delivery":Expected_Delivery, "Availability":Availability, "URL":urls})
df
```

# Q3: Write a python program to access the search bar and search button on images.google.com and scrape 10 images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’, ‘Guitar’, ‘Cakes’.¶

```python
#connecting with the webdriver

driver = webdriver.Chrome()
```

```python
#opening the google images website on automated chrome browser

url="http://images.google.com/"
driver.get(url)
```

```python
# creating empty list
urls = []
data = []

search_item = ["Fruits", "Cars", "Machine Learning", "Guitar", "Cakes"]
for item in search_item:
    driver.get(url)
    time.sleep(5)
    
    # clicking on the search_bar
    search_bar = driver.find_element(By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/textarea')
    
    # sending the search items to the search bar
    search_bar.send_keys(str(item))
    
    # clicking on search button
    search_button = driver.find_element(By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/button/div/span')
    search_button.click()
    
    # scroling down the webpage to get some more images
    for _ in range(50):
        driver.execute_script("window.scrollBy(0,100)")
        
        imgs = driver.find_elements(By.XPATH, '//div[@class="fR600b islir"]')
    img_url = []
    for image in imgs:
        source = image.get_attribute('src')
        if source is not None:
            if(source[0:4] == 'http'):
                img_url.append(source)
    for i in img_url[:10]:
        urls.append(i)
        
for i in range(len(urls)):
    if i >= 30:
        break
    print("Downloading {0} of {1} images" .format(i,30))
    response = requests.get(urls[i])
    
    file = open(r"E:\google\images"+str(i)+".jpg","wb")
    
    file.write(response.content)
```

# Q5: Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps.

```python
#connecting with the webdriver

driver = webdriver.Chrome()
```

```python
#opening the google maps website on automated chrome browser

driver.get("https://www.google.co.uk/maps/")
```

```python
# take an input from the user

user_input = input('Search for city: ')
```

```python
#searching for the user input in the search field

search_item=driver.find_element(By.XPATH, '/html/body/div[3]/div[8]/div[3]/div[1]/div[1]/div/div[2]/form/input')
search_item.send_keys(user_input)
```

```python
search_button = driver.find_element(By.XPATH, '/html/body/div[3]/div[8]/div[3]/div[1]/div[1]/div/div[2]/div[1]/button')
search_button.click()
```

```python
try:
    url_str = driver.current_url
    latitude_longitude = re.findall(r'@(.*),',url_str)
    #print(latitude_longitude)
    if len(latitude_longitude):
        lat_lng_list = latitude_longitude[0].split(",")
        if len(lat_lng_list)>=2:
            latitude = lat_lng_list[0]
            longitude = lat_lng_list[1]
        print("Latitude = {}, Longitude = {}".format(latitude, longitude))

except Exception as e:
        print("Error: ", str(e))
```

# Q9:Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in “London” location.
```python
driver=webdriver.Chrome()
```

```python
driver.get("https://www.hostelworld.com/")
```

```python
accept_cookies=driver.find_element(By.XPATH, '/html/body/div[3]/div/div/div[4]/div/div/div[2]/button[2]')
```

```python
location = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div/div/div[1]/div[1]/div/div[2]/input')
location.send_keys('London')
```

```python
drop_list=driver.find_element(By.XPATH, '/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div/div/div[1]/div[2]/div/ul/li[2]/button/div[2]/div[1]')
drop_list.click()
```

```python
search_button=driver.find_element(By.XPATH, '/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div/div/div[5]/button[1]')
search_button.click()
```

```python
hostel_name=[]
distance=[]
rating=[]
total_reviews=[]
review=[]
price=[]



start=0
end=2

for page in range(start,end):
    name_tags=driver.find_elements(By.XPATH,'//div[@class="property-name"]')[1:]
    for i in name_tags:
        hostel_name.append(i.text)

    
    distance_tags=driver.find_elements(By.XPATH,'//span[@class="distance-description"]')
    for i in distance_tags:
        distance.append(i.text)
    
    rating_tags=driver.find_elements(By.XPATH,'//span[@class="number"]')[1:]
    for i in rating_tags:
        rating.append(i.text)
    
    total_tags=driver.find_elements(By.XPATH, '//span[@class="left-margin"]')
    for i in total_tags:
        total_reviews.append(i.text)
           
    review_tags=driver.find_elements(By.XPATH,'//span[@class="keyword"]')[1:]
    for i in review_tags:
        review.append(i.text)
        
    price_tags=driver.find_elements(By.XPATH,'//div[@class="property-accommodation-prices"]')
    for i  in price_tags:
        price.append(i.text)
        
        
        
    
    
```

```python
print(len(hostel_name), len(distance), len(rating), len(total_reviews), len(review), len(price))
```

```python
urls=[]
start=0
end=2

for page in range(start,end):
    url=driver.find_elements(By.XPATH,'//a[@class="property-card-container horizontal"]')
    for i in url:
        urls.append(i.get_attribute('href'))
    next_button=driver.find_elements(By.XPATH, '/html/body/div[3]/div/div/div[2]/div[2]/div[1]/div/div/div[5]/div/button[2]')
```

```python
facilities=[]
description=[]

for p in urls:
    driver.get(p)
    time.sleep(3)
    try:
        facility=driver.find_element(By.XPATH,'//ul[@class="facility-sections"]')
        facilities.append(facility.text.split('|'))
    except NoSuchElementException :
        facilities.append('-')
        
    
    try:
        desc=driver.find_element(By.XPATH,'//div[@class="content collapse-content"]')
        description.append(desc.text)
    except NoSuchElementException :
        description.append('-')

print(len(facilities), len(description))
```

```python
df=pd.DataFrame({"Hostel Name":hostel_name, "Distance":distance, "Rating":rating,"Total Reviews":total_reviews, "Overal reviews":review, "Price":price, "Facilities":facilities, "Description":description})
df
```
